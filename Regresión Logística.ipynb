## 1. Realizar un análisis exploratorio de los datos para identificar relaciones entre variables, valores atípicos, tendencias, etc.

```python
# lectura de datos en Python y lo cargamos en la variable Datos

Datos = pd.read_csv('./heart_cleveland_upload.csv')
```

## 2. Preprocesar los datos limpiándolos, tratando valores faltantes y transformándolos según sea necesario

```python
# Revisión de datos faltantes en valores no binarios y no categóricos

print("Cantidad de Registros con cero en Edad: ", Datos['age'].isin([0]).sum())
print("Cantidad de Registros con cero en Presión Arterial: ", Datos['trestbps'].isin([0]).sum())
print("Cantidad de Registros con cero en Colesterol: ", Datos['chol'].isin([0]).sum())
print("Cantidad de Registros con cero en Frecuencia Cardíaca Máxima: ", Datos['thalach'].isin([0]).sum())
```

```python
#Identificación de Datos Atípicos

plt.figure(figsize=(6, 3))
sns.boxplot(x=Datos['thalach'])
plt.title('Identificación de Datos Atípicos', fontsize=10 )
```

```python
#Conteo de Datos Atípicos

nivel_minimo=60
nivel_maximo=180
Total_Atipicos=((Datos['thalach']<nivel_minimo)|(Datos['thalach']>nivel_maximo)).sum()
print("El total de datos atípicos es {}".format(Total_Atipicos))
```

```python
# Filtra Datos Atípicos

Datos=Datos[(Datos['thalach']<nivel_maximo)&(Datos['thalach']>nivel_minimo)]
```

```python
# Validar si se removieron los datos atípicos

Total_Atipicos=((Datos['thalach']<nivel_minimo)|(Datos['thalach']>nivel_maximo)).sum()
print("El total de datos atípicos es {}".format(Total_Atipicos))
```

```python
#Revisión del tipo de datos y datos faltantes

Datos.info()
```

## 3.Seleccionar las características más relevantes para entrenar el modelo utilizando selección de características

Característica de interes:
* Condición: si la persona tiene condición de enfermedad cardiaca o no (0 = no tiene condición - 1 = si tiene condición)
* Predecir si una persona puede tener enfermedad cardiaca

```python
# Total de personas con Diabetes

Datos['condition'].value_counts()
```

```python
Datos[['condition']].head(40)
```

## 4. Dividir el dataset en Train y Test para evaluar correctamente el modelo

```python
# Porcentaje de datos para entrenamiento (train) = 75% , y para prueba (test) = 25%

# Se crean variables X y Y para almacenar la información del entrenamiento

X = Datos.drop('condition',axis=1) # Se cargan los datos de train sin los datos de la columna Diabetico 
Y = Datos['condition'] # Se cargan los datos de train con los datos de la la columna Diabetico

#La función train_test_split crea una división de un conjunto de datos en dos bloques uno de entrenamiento y otro de prueba (train and test)
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.25, stratify=Y, random_state= 0)
```

```python
print (X_train)
```

```python
print (Y_train)
```

## 5. Entrenar el modelo configurando los diferentes hiperparámetros

```python
# Construir el modelo de regresión logística 

modelo_regresion = LogisticRegression(solver = "liblinear") # liblinear es una libreria para resolver problemas lineales de clasificación y regresión
modelo_regresion.fit (X_train , Y_train)   

# Crea el modelo de prediccion
Prediccion = modelo_regresion.predict(X_test) # la función predict predice los valores a partir del conjunto de datos de validación
```

```python
print (Y_test)
```

```python
print (Prediccion)
```

## 6. Evaluar el desempeño del modelo en el conjunto de Test con métricas como precisión, recall, F1-score, etc

```python
# Desempeño del modelo de regresión

print(classification_report(Y_test,Prediccion))
```

## 7. Realizar las diferentes gráficas que permitan visualizar los resultados del modelo

```python
#Matrix de Confusión del modelo

cm=confusion_matrix(Y_test,Prediccion)
disp=metrics.ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot()
```

```python
# Curva de precisón del modelo

from sklearn.metrics import RocCurveDisplay
RocCurveDisplay.from_predictions(Y_test, Prediccion)
plt.show()
```

## 8. Interpretar, analizar y documentar los resultados obtenidos

Del desempeño del modelo se puede describir:

- Precisión: Precisión mide la proporción de predicciones positivas verdaderas entre todas las predicciones positivas realizadas por el modelo. Aquí la precisión para la clase "condition" es 0,88, lo que significa que el 88% de los predichos como condicion de enfermedad cardiaca por el modelo eran en realidad enfermos cardiacos.

- Recall: Mide la proporción de predicciones positivas verdaderas entre todas las instancias positivas reales en el conjunto de datos. Aquí el Recall para la clase "condition" es 0,65, lo que indica que el modelo identificó correctamente al 65% de todos los enfermos cardiacos.

- F1 Score: Es la media armónica de precisión y recuperación. Proporciona una puntuación única que equilibra la Precisión y Recall. Aquí la puntuación F1 para la clase "condition" es 0,75, lo que indica un equilibrio entre precisión y recuerdo.


De la Matriz de Confusión podemos interpretar:

- Verdaderos negativos: el recuento de observaciones que un clasificador predijo correctamente como falso 33

- Verdaderos positivos: el recuento de observaciones que un clasificador predijo correctamente como Verdadero 22

- Falsos positivos: el recuento de observaciones que un clasificador predijo incorrectamente como Verdaderas 3

- Falsos negativos: el recuento de observaciones que un clasificador predijo incorrectamente como falso 12

De la curva de precisión del modelo se puede interpretar:

- El AUC es 0,78 lo que significa que el 78 % de las predicciones del clasificador son correctas en todos los umbrales de clasificación.
